## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset "diamonds" do Seaborn
diamonds = sns.load_dataset("diamonds")
```

```
# Criar o gráfico de barras para preço médio por tipo de corte
plt.figure(figsize=(10, 6))
sns.barplot(data=diamonds, x="cut", y="price", palette="viridis", estimator="mean")
plt.title("Preço Médio do Diamante por Tipo de Corte", fontsize=16)
plt.xlabel("Tipo de Corte", fontsize=12)
plt.ylabel("Preço Médio (USD)", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Verificar o peso médio (carat) por tipo de corte
mean_carat_by_cut = diamonds.groupby("cut")["carat"].mean()
print("Peso médio (carat) por tipo de corte:")
print(mean_carat_by_cut)

# Gráfico 2: Distribuição do peso (carat) por tipo de corte
plt.figure(figsize=(10, 6))
sns.boxplot(data=diamonds, x="cut", y="carat", palette="coolwarm")
plt.title("Distribuição do Peso (Carat) por Tipo de Corte", fontsize=16)
plt.xlabel("Tipo de Corte", fontsize=12)
plt.ylabel("Peso (Carat)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

**Insight do gráfico 1**: ...

Insights e Explicações:

Gráfico 1: Preço Médio do Diamante por Tipo de Corte

Insight:

O tipo de corte "Fair" apresenta o maior preço médio, enquanto o tipo "Ideal" tem o menor preço médio.
Isso pode parecer contraintuitivo, já que cortes de maior qualidade, como "Ideal" e "Premium", geralmente são associados a preços mais altos.
Motivo da Verificação do Peso (Carat):

O peso do diamante (medido em carat) é um fator significativo no preço de um diamante.
Para entender por que diamantes com corte "Fair" têm o maior preço médio, foi necessário verificar se esses diamantes possuem pesos médios maiores.

Peso Médio (Carat) por Tipo de Corte
Insight:

O peso médio dos diamantes varia entre os tipos de corte:
"Fair" tem o maior peso médio (aproximadamente 1.04 carat).
"Ideal" tem o menor peso médio (aproximadamente 0.70 carat).
Isso explica por que diamantes com corte "Fair" têm um preço médio maior: eles tendem a ser significativamente mais pesados.


Gráfico de Boxplot:

O gráfico mostra a distribuição do peso para cada tipo de corte.
Cortes como "Fair" têm pesos medianos e máximos maiores, enquanto cortes como "Ideal" têm pesos menores.


Conclusão

O preço médio do diamante é influenciado não apenas pelo tipo de corte, mas também pelo peso (carat).

Embora cortes como "Ideal" e "Premium" sejam de maior qualidade, diamantes "Fair" frequentemente têm pesos maiores, o que aumenta o preço médio.

Adicionar a análise do peso foi essencial para compreender as discrepâncias entre a qualidade do corte e o preço médio.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# Filtrar para dados de dezembro
december_data = data[data["month"] == "Dec"]
print(december_data)
```

```
# Filtrar os dados para o mês de dezembro
december_data = data[data["month"] == "Dec"]

# Criar o gráfico de barras para passageiros em dezembro por ano
plt.figure(figsize=(10, 6))
sns.barplot(data=december_data, x="year", y="passengers", palette="viridis")
plt.title("Número de Passageiros em Dezembro por Ano", fontsize=16)
plt.xlabel("Ano", fontsize=12)
plt.ylabel("Número de Passageiros", fontsize=12)
plt.show()
```

**Insight do gráfico 2**: ...

Gráfico 2: Número de Passageiros em Dezembro por Ano
Crescimento Sustentado:

O gráfico indica um crescimento constante no número de passageiros em dezembro entre 1949 e 1960.
Esse crescimento pode refletir o aumento da popularidade das viagens aéreas ao longo dos anos devido à expansão econômica ou maior acessibilidade.
Possível Influência Sazonal:

O mês de dezembro pode apresentar um aumento no número de passageiros devido a fatores sazonais, como as férias de final de ano e feriados.
Tendência Linear:

O aumento é quase linear, sugerindo que o setor de aviação crescia de forma constante na época, possivelmente devido à introdução de mais rotas e aeronaves com maior capacidade.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# Filtrar o conjunto de dados entre 1949 e 1959
filtered_data = data[(data["year"] >= 1949) & (data["year"] <= 1959)]

# Criar o gráfico de linhas para o número de passageiros por mês entre 1949 e 1959
plt.figure(figsize=(14, 8))
sns.lineplot(data=filtered_data, x="month", y="passengers", hue="year", marker="o", palette="tab10")
plt.title("Número de Passageiros por Mês (1949-1959)", fontsize=16)
plt.xlabel("Mês", fontsize=12)
plt.ylabel("Número de Passageiros", fontsize=12)
plt.xticks(rotation=45)  # Melhor visualização dos rótulos dos meses
plt.legend(title="Ano", bbox_to_anchor=(1.05, 1), loc='upper left')  # Posicionamento da legenda
plt.show()
```

**Insight do gráfico 3**: ...

Gráfico 3: Número de Passageiros por Mês entre 1949 e 1959
Padrão Sazonal:

O número de passageiros atinge picos consistentes em julho e agosto, indicando que esses meses provavelmente coincidem com períodos de férias, viagens a lazer e maior demanda.
Em contraste, os meses de inverno (como janeiro e fevereiro) têm menor número de passageiros.
Crescimento Anual:

Apesar da sazonalidade, existe um aumento geral no número de passageiros ao longo dos anos.
Em 1949, o pico em agosto está abaixo de 400 passageiros, enquanto em 1959, esse pico ultrapassa 600 passageiros.
Diferença Consistente entre Meses:

A diferença entre o menor e o maior número de passageiros ao longo de cada ano permanece estável, destacando que os padrões sazonais se mantêm constantes apesar do crescimento geral.

Conclusão

Ambos os gráficos mostram o crescimento constante da aviação durante o período analisado.
O gráfico 2 destaca o aumento linear ao longo dos anos para um único mês (dezembro), enquanto o gráfico 3 evidencia sazonalidade e picos consistentes nos meses de verão.
Essas tendências são úteis para:
Planejamento de capacidade por parte das companhias aéreas.
Identificação de períodos de alta demanda para estratégias de marketing e otimização de preços.
