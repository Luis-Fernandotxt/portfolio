## 1\. Preços `outliers` de diamante

```
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset "diamonds"
diamonds = sns.load_dataset("diamonds")
```

```
# Criar um boxplot para identificar outliers no preço dos diamantes
plt.figure(figsize=(10, 6))
sns.boxplot(data=diamonds, x="price", color="skyblue")
plt.title("Boxplot dos Preços de Diamantes", fontsize=16)
plt.xlabel("Preço (USD)", fontsize=12)
plt.show()
```

**Insight do gráfico 1**: ...

Gráfico 1: Boxplot dos Preços de Diamantes

**Insight Principal:** A maior parte dos preços dos diamantes está concentrada em uma faixa de até **10.000 USD**, com mediana próxima de **4.000 USD**. Foram identificados **outliers significativos** acima de **12.500 USD**, indicando diamantes com características excepcionais (como peso maior ou qualidade superior).

**Interpretação:** Esses outliers podem ser investigados mais a fundo para entender quais fatores (peso, transparência ou corte) os tornam tão valiosos. O gráfico demonstra uma ampla dispersão de preços, sugerindo que diferentes fatores influenciam significativamente o valor do diamante.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# Criar um gráfico de dispersão para mostrar a correlação entre preço e peso, agrupados por transparência
plt.figure(figsize=(12, 8))
sns.scatterplot(data=diamonds, x="carat", y="price", hue="clarity", palette="tab10", alpha=0.7)
plt.title("Correlação entre Preço e Peso (Agrupado por Transparência)", fontsize=16)
plt.xlabel("Peso (Carat)", fontsize=12)
plt.ylabel("Preço (USD)", fontsize=12)
plt.legend(title="Transparência (Clarity)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Insight do gráfico 2**: ...

Gráfico 2: Correlação entre Preço e Peso (Agrupado por Transparência)

**Insight Principal:** Há uma **forte correlação positiva** entre o peso (**carat**) e o preço do diamante. Diamantes mais pesados tendem a ser mais caros. A transparência (**clarity**) afeta os preços dentro de cada faixa de peso: diamantes com transparência **IF (Internally Flawless)** e **VVS1** possuem valores significativamente mais altos em relação a categorias menos claras como **SI2** e **I1**.

**Interpretação:** O gráfico destaca a importância combinada do peso e da transparência no preço dos diamantes. Enquanto o peso é o principal fator, a transparência atua como um multiplicador de valor. Há uma grande densidade de dados em faixas de peso menores (até **2 carats**), indicando que a maioria dos diamantes comercializados está nesse intervalo.

Gráfico 3: Distribuição Contínua Aproximada do Peso dos Carros

**Insight Principal:** A distribuição do peso dos carros é **assimétrica à direita**, com a maioria dos carros concentrados na faixa de peso entre **2.000 e 3.000 lbs**. Há uma queda acentuada na densidade para carros mais leves (< **2.000 lbs**) e mais pesados (> **4.000 lbs**), indicando que esses são menos comuns.

**Interpretação:** Esse padrão reflete os padrões típicos de fabricação de carros, com a maioria dos modelos produzidos em um intervalo de peso ideal para eficiência e desempenho. A presença de alguns carros muito pesados (> **5.000 lbs**) pode indicar veículos utilitários ou de luxo, que são menos comuns no mercado geral.
